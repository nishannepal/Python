# Cooks Distance

```python
from yellowbrick.regressor import CooksDistance
import numpy as np
import pandas as pd
```

```python
converted_file_prefix = 'new_lasso'
df = pd.read_excel('new_lasso.xlsx')
df.to_csv(converted_file_prefix + ".csv")
data = pd.read_csv('new_lasso.csv')
```

```python
X = data.iloc[:,2:7].values
y = data.iloc[:,7].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X , y
```

```python
from yellowbrick.datasets import load_concrete
from yellowbrick.regressor import cooks_distance

visualizer = CooksDistance()
visualizer.fit(X, y)
visualizer.show(outpath="new_Cook's_distance.png")
```

# Pearson's Correlation heatmap after eliminating outliers

```python
converted_file_prefix = 'new_cook_corr'
df = pd.read_excel('newfeat_corr_lasso_cook.xlsx')
df.to_csv(converted_file_prefix + ".csv")
data = pd.read_csv('new_cook_corr.csv')
```

```python
X = data.iloc[:,2:7]
y = data.iloc[:,7]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,5))
cor = X.corr()
sns.heatmap(cor, annot = True, cmap = plt.cm.CMRmap_r,  annot_kws={"size": 13})
plt.savefig('newcorr90.png')
plt.show()
```
